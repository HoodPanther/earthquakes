```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.style.use('ggplot')
```

```python
# earthquakes dataframe
earthquakes = pd.read_csv('all_month.csv')
earthquakes.head()
```

```python
# tsunamis dataframe
tsunamis = pd.read_excel('tsevent.xlsx')
tsunamis.head()
```

```python
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import rgb2hex
from matplotlib.patches import Polygon
```

```python
# delete unnecessary columns
tsunamis.drop(tsunamis.columns[[range(22,46)]], inplace = True, axis = 1)
```

```python
tsunamis.head()
```

```python
print earthquakes['mag'].max(), earthquakes['mag'].min()
```

```python
# small earthquake df
smallearthquakes = earthquakes.loc[earthquakes['mag'] < 2.0]
smallearthquakes.head()
```

```python
len(smallearthquakes.index)
```

```python
# big earthquake df
bigearthquakes = earthquakes.loc[earthquakes['mag'] > 5.0]
bigearthquakes.head()
```

```python
len(bigearthquakes.index)
```

```python
# Drop N/A lon/lat values for tsunami
# I filtered with longitude because if longitude has N/A, corresponding latitude also has it
tsu = tsunamis.loc[np.isnan(tsunamis[u'LONGITUDE']) == False]
tsu
```

I dropped the observations with longitudes and latitudes that don't have values otherwise, I wouldn't be able to plot those observations on a world map.

```python
# draw world map
plt.figure(figsize=(15,10))
displaymap = Basemap(llcrnrlon=-180,llcrnrlat=-90,urcrnrlon=180,urcrnrlat=90)
```

```python
displaymap.drawmapboundary()
displaymap.drawcountries()
displaymap.drawcoastlines()
```

```python
# Convert longitudes and latitudes to list of floats
longitude = earthquakes[['longitude']].values.tolist()
for i in range(0, len(longitude)):
    longitude[i] = float(longitude[i][0])
latitude = earthquakes[['latitude']].values.tolist()
for i in range(0, len(latitude)):
    latitude[i] = float(latitude[i][0])
tlongitude = tsu[[u'LONGITUDE']].values.tolist()
for i in range(0, len(tlongitude)):
    tlongitude[i] = float(tlongitude[i][0])
tlatitude = tsu[[u'LATITUDE']].values.tolist()
for i in range(0, len(tlatitude)):
    tlatitude[i] = float(tlatitude[i][0])
```

```python
lons,lats = displaymap(longitude, latitude)
tlons, tlats = displaymap(tlongitude, tlatitude)
displaymap.plot(lons, lats, 'bo', color = "blue")
displaymap.plot(tlons, tlats, 'bo', color = "red")
```

```python
plt.show()
```

In this map, I plotted all the known points for the earthquakes dataset and all the known points for the tsunami datasets. Some of the points in the North American region and the east Asia region seem to overlap so it is implied that there may be some correlation between earthquakes and tsunamis there. 

```python
plt.figure(figsize=(15,10))
slongitude = smallearthquakes[['longitude']].values.tolist()
for i in range(0, len(slongitude)):
    slongitude[i] = float(slongitude[i][0])
slatitude = smallearthquakes[['latitude']].values.tolist()
for i in range(0, len(slatitude)):
    slatitude[i] = float(slatitude[i][0])
displaymaps = Basemap(llcrnrlon=-180,llcrnrlat=-90,urcrnrlon=180,urcrnrlat=90)
slons,slats = displaymap(slongitude, slatitude)
displaymaps.plot(slons, slats, 'bo', color = "blue")
displaymaps.plot(tlons, tlats, 'bo', color = "red")
displaymaps.drawmapboundary()
displaymaps.drawcountries()
displaymaps.drawcoastlines()
```

```python
plt.show()
```

In this map, I plotted all the known points all small earthquakes and all the known points for the tsunami datasets. Some of the points in the North American region seem to overlap but most of the datapoints for small earthquakes are in the North American region which could be the cause of some correlation between small earthquakes and tsunamis.

```python
plt.figure(figsize=(15,10))
blongitude = bigearthquakes[['longitude']].values.tolist()
for i in range(0, len(blongitude)):
    blongitude[i] = float(blongitude[i][0])
blatitude = bigearthquakes[['latitude']].values.tolist()
for i in range(0, len(blatitude)):
    blatitude[i] = float(blatitude[i][0])
displaymapb = Basemap(llcrnrlon=-180,llcrnrlat=-90,urcrnrlon=180,urcrnrlat=90)
blons,blats = displaymap(blongitude, blatitude)
displaymapb.plot(blons, blats, 'bo', color = "blue")
displaymapb.plot(tlons, tlats, 'bo', color = "red")
displaymapb.drawmapboundary()
displaymapb.drawcountries()
displaymapb.drawcoastlines()
```

```python
plt.show()
```

In this map, I plotted all the known points all big earthquakes and all the known points for the tsunami datasets. There does not seem to be any visible correlation between the two but there aren't a lot of big earthquakes observation in the dataset which may be the cause of this. If there's more data to go on for big earthquakes, I may be able to get a better picture of the correlation between big earthquakes and tsunamis.

One of the factors that could cause errors in my observations is that the data from the earthquakes are from recent times while the dataset for tsunamis comes from throughout history so this may not be an accurate representation of correlation between earthquakes and tsunamis. More data on earthquakes, especially data on earthquakes throughout history would be better to get a more accurate representation of the correlation.
